```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('spotify_data.csv')
data.head()
```

```python
data.info()
```

```python
data = data.dropna().reset_index(drop=True)
data.info()
```

```python
data.tail()
```

```python
# remove most categorical columns to extract mainly the audio features
data.drop(columns=['Unnamed: 0','track_id','album_name','track_name'], inplace=True)
data.info()
```

```python
data['duration_ms'] = data['duration_ms']/1000      # convert duration from ms to seconds
data = data.rename(columns={'duration_ms':'duration'})
data['explicit'] = data['explicit'].astype(int)     # convert boolean values to 0 and 1 respectively

data.info()
```

```python
data.head()
```

```python
# extract kpop genre songs
kpop_data = data[data['track_genre']=='k-pop'].reset_index(drop=True)
kpop_data
```

```python
kpop_data['artists'].nunique()
```

```python
# let's remove all rows indicating collaborations (i.e. multiple artists)
kpop_data_filt = kpop_data[~kpop_data['artists'].str.contains(';|&')]
kpop_data_filt = kpop_data_filt.reset_index(drop=True)
kpop_data_filt
```

```python
kpop_data_filt['artists'].nunique()
```

```python
kpop_data_filt['artists'].value_counts()
```

```python
# remove non K-pop/Korean artists
non_korean = ['T. M. Soundararajan', 'Vijay Antony', 'Yuvan Shankar Raja','Alka Yagnik', 
              'All Ok', 'Anupam Roy', 'Anuradha Paudwal', 'Dhanush', 'Ilaiyaraaja', 'Jack Harlow',
              'Jagjit Singh','Jubin Nautiyal', 'Madhusree', 'ONE OK ROCK', 'Prem', 'RADWIMPS', 
              'Raghu Dixit', 'Sangeetha Rajeev', 'Shachina Heggar', 'Siddhi Ture']

kpop_data_filt = kpop_data_filt[~kpop_data_filt['artists'].isin(non_korean)]
kpop_data_filt = kpop_data_filt.reset_index(drop=True)
kpop_data_filt
```

```python
print("Number of Kpop artists: ",kpop_data_filt['artists'].nunique())
kpop_data_filt['artists'].value_counts()
```

```python
print("Average number of times Kpop artists are featured: ", kpop_data_filt['artists'].value_counts().mean())
```

```python
kpop_artists_5_times = (kpop_data_filt['artists'].value_counts() >= 5).sum()
print("Number of artists that are featured at least 5 times: ", kpop_artists_5_times)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,12))
sns.countplot(data=kpop_data_filt, x='artists')
plt.xticks(rotation=90)
plt.xlabel('Artist')
plt.ylabel('Number of songs')
plt.show()
```

Artists with the most number of popular songs are the Kpop groups. 

```python
# for ease, we will group artists with songs less than 5 as 'others' 
# defining a general function
def shorten_categories(categories, cutoff):
    categorical_map = {}
    for i in range(len(categories)):
        if categories.values[i]>=cutoff:
            categorical_map[categories.index[i]] = categories.index[i]
        else:
            categorical_map[categories.index[i]] = 'Others'
            
    return categorical_map
```

```python
artists_map = shorten_categories(kpop_data_filt.artists.value_counts(), 5)
kpop_data_filt['artists'] = kpop_data_filt['artists'].map(artists_map)

kpop_data_filt.artists.value_counts()
```

```python
# popularity vs artist relationship
plt.figure(figsize=(15,12))
sns.boxplot(data=kpop_data_filt, x='artists',y='popularity')
plt.xticks(rotation=90)
plt.xlabel('')
plt.ylabel('Popularity')
plt.ylim([40,100])
plt.title('Artist Popularity')
plt.show()
```

Blackpink has the widest range of popularity levels and owns the highest recorded popularity score of about 87-88.

```python
# since we only have kpop artists (single genre), we do not need this column anymore
kpop_data_filt.drop('track_genre', axis=1, inplace=True)
kpop_data_filt.columns
```

```python
kpop_data_filt.to_csv('kpop_nonencoded_data.csv')
```

**one-hot encoding of kpop artist names**

```python
def one_hot_encode(df, column, prefix):
    df = df.copy()
    dummies = pd.get_dummies(df[column], prefix=prefix)
    df = pd.concat([df, dummies], axis=1)
    df = df.drop(column, axis=1)
    return df
```

```python
kpop_data_filt = one_hot_encode(kpop_data_filt, 'artists', 'artist')

kpop_data_filt
```

```python
# check the final dataset
kpop_data_filt.info()
```

**Numerical correlations**

```python
plt.figure(figsize=(12,8))
sns.heatmap(kpop_data_filt.corr(), vmin=-1, vmax=1, cmap='RdBu')
plt.show()
```

```python
kpop_data_filt.corr()
```

We see that some artists are positively correlated with (e.g. Twice and Stray Kids) and some artists are negatively correlated with popularity (e.g. BTS, Jeon Somi).


Popularity is also positively correlated with how upbeat or positive a song is (e.g. valence, energy, danceability, loudness) and having explicit language in the song is also positively correlated with popularity.
